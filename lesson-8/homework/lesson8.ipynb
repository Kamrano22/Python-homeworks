```python
with open('my_file.txt', 'r') as f:
    words = f.read().split()  # read file and split into words

max_len = max(len(word) for word in words)
longest_words = [word for word in words if len(word) == max_len]

print("Longest word length:", max_len)
print("Longest word(s):", longest_words)
```

```python
with open('my_file.txt', 'r') as f:
    text = f.read().lower()
    words = text.split()

word_counts = {}

for word in words:
    if word in word_counts:
        word_counts[word] += 1
    else:
        word_counts[word] = 1

print(word_counts)
```

```python
import string

for letter in string.ascii_uppercase:  # 'A' to 'Z'
    filename = f"{letter}.txt"
    with open(filename, 'w') as f:
        f.write(f"This is file {filename}\n") 

        print(filename)
```

```python
import string

letters = string.ascii_lowercase  # all letters 'a' to 'z'

letters_per_line = 5  # specify how many letters you want per line

with open('alphabet.txt', 'w') as f:
    for i in range(0, len(letters), letters_per_line):
        line = letters[i:i+letters_per_line]
        f.write(line + '\n')


        print(line)
```

```python
filenames = ['my_file.txt', 'destination.txt']  # replace with your file names
all_chars = []

for filename in filenames:
    with open(filename, 'r', encoding='utf-8') as f:
        content = f.read()
        all_chars.extend(list(content))  # this uses the built-in list function

print(all_chars)
```
