```python
class ToDoList:
    def __init__(self):
        self.tasks = []

    def add_task(self, task):
        self.tasks.append({"task": task, "completed": False})

    def remove_task(self, task):
        self.tasks = [t for t in self.tasks if t["task"] != task]

    def complete_task(self, task):
        for t in self.tasks:
            if t["task"] == task:
                t["completed"] = True

    def view_tasks(self):
        for t in self.tasks:
            status = "Done" if t["completed"] else "Pending"
            print(f"- {t['task']} [{status}]")



todo = ToDoList()
todo.add_task("Buy groceries")
todo.add_task("Study Python")
todo.view_tasks()

todo.complete_task("Buy groceries")
todo.view_tasks()

todo.remove_task("Study Python")
todo.view_tasks()


def print_menu():
    print("\nToDoList management System menu:")
    print('1.add_task')
    print('2.remove_task')
    print('3.complete_task')
    print('4.view_tasks')
    print('5.reprint the menu')
    print('6.exit')

todo = ToDoList()
print_menu()

while True:
    choice = input("\nEnter your choice (1-6): ").strip()

    if choice == '1':
        task = input("Enter the task to add: ")
        todo.add_task(task)
        print(f"Task added: {task}")

    elif choice == '2':
        task = input("Enter the task to remove: ")
        todo.remove_task(task)
        print(f"Task removed: {task}")

    elif choice == '3':
        task = input("Enter the task to mark as completed: ")
        todo.complete_task(task)
        print(f"Task marked as completed: {task}")

    elif choice == '4':
        print("\nYour Tasks:")
        todo.view_tasks()

    elif choice == '5':
        print_menu()

    elif choice == '6':
        print("Exiting ToDoList. Goodbye!")
        break

    else:
        print("Invalid input. Please enter a number from 1 to 6.")
```

```python
class Account:
    def __init__(self, account_number, owner, balance):
        self.acc_id = account_number
        self.owner = owner
        self.balance = balance

class Bank:
    def __init__(self):
        self.accounts = {}

    def add_account(self, name, balance=0):
        self.accounts[name] = balance

    def check_balance(self, name):
        if name in self.accounts:
            print(f"{name}'s balance is: ${self.accounts[name]}")
        else:
            print("Account not found.")

    def deposit(self, name, amount):
        if name in self.accounts:
            self.accounts[name] += amount
            print(f"${amount} deposited. New balance: ${self.accounts[name]}")
        else:
            print("Account not found.")

    def withdraw(self, name, amount):
        if name in self.accounts:
            if self.accounts[name] >= amount:
                self.accounts[name] -= amount
                print(f"${amount} withdrawn. New balance: ${self.accounts[name]}")
            else:
                print("Insufficient balance.")
        else:
            print("Account not found.")


def print_menu():
    print('\nAccount Management System:')
    print('1. Add Account')
    print('2. Check Balance')
    print('3. Deposit')
    print('4. Withdraw')
    print('5. Exit')


# Main program starts here
my_account = Bank()
print_menu()

while True:
    choice = input('Enter your choice (1-5): ').strip()

    if choice == '1':
        name = input("Enter account name: ")
        initial = float(input("Enter initial deposit: "))
        my_account.add_account(name, initial)
        print(f"Account '{name}' added with ${initial}.")

    elif choice == '2':
        name = input("Enter account name to check balance: ")
        my_account.check_balance(name)

    elif choice == '3':
        name = input("Enter account name to deposit to: ")
        amount = float(input("Enter deposit amount: "))
        my_account.deposit(name, amount)

    elif choice == '4':
        name = input("Enter account name to withdraw from: ")
        amount = float(input("Enter withdrawal amount: "))
        my_account.withdraw(name, amount)

    elif choice == '5':
        print("Exiting the bank system. Goodbye!")
        break

    else:
        print("Invalid input. Please choose a number between 1 and 5.")


        

                    
                    

            

    










```
